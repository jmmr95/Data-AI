**Descripción del conjunto de datos.**

Este conjunto de datos contiene información sobre productos electrónicos disponibles para venta en línea, enfocados en la categoría Audio y Video. Cada registro representa un artículo listado en una plataforma de ecommerce y cuenta con detalles sobre su descripción, precio, popularidad, disponibilidad de envío y fecha de publicación.


**Columnas principales:**

Product Category: Categoría principal del producto (todos del tipo "Audio Video").

Product Description: Descripción comercial detallada del producto.

Price (Dollar): Precio en dólares del artículo.

Number of Reviews: Indicador de popularidad basado en la cantidad de reseñas.

Shipment: Información sobre disponibilidad o destino de envío (en este caso, todos se envían a Bangladesh).

Order Date: Fecha de publicación o disponibilidad del producto en la tienda online.

**Este dataset permite realizar análisis exploratorios como:**

Comparación de precios y popularidad entre productos.

Identificación de productos más vendidos o mejor valorados.

Análisis de tendencias de lanzamiento o stock a lo largo del tiempo.

Estudio del comportamiento del mercado para dispositivos electrónicos portátiles.
