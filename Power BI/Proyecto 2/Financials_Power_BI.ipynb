**Descripción del conjunto de datos**

Este conjunto de datos representa transacciones de ventas de productos a nivel internacional, segmentadas por tipo de cliente, país y fecha. Cada registro refleja una operación comercial con detalles sobre unidades vendidas, precios, ingresos, costos y márgenes de ganancia.

**Columnas principales:**

Segment: Tipo de cliente o mercado (e.g., Government, Midmarket).

Country: País donde se realizó la venta.

Product: Tipo de producto vendido (e.g., Carretera).

Discount Band: Rango de descuento aplicado.

Units Sold / Sale Price / Gross Sales / Discounts / Sales: Métricas de ingresos y descuentos aplicados.

COGS (Cost of Goods Sold): Costo directo de los productos vendidos.

Profit: Ganancia obtenida por transacción.

Date / Month / Year: Variables temporales para análisis estacional.

**Este dataset permite realizar análisis centrados en:**

Rentabilidad por segmento y país.

Comportamiento de ventas a lo largo del tiempo.

Impacto de los descuentos en los ingresos netos.

Comparación entre ingresos brutos, descuentos y utilidades.
