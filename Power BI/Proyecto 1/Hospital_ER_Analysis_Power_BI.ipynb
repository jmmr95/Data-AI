**Descripción del conjunto de datos hospitalarios**

Este conjunto de datos contiene información relacionada con la admisión y atención de pacientes en un hospital. Cada fila representa un paciente admitido, e incluye atributos demográficos, detalles de su ingreso, área de referencia médica, experiencia del paciente y tiempos de espera.

**Columnas principales:**

Patient Id: Identificador único del paciente (anonimizado).

Admission Date: Fecha y hora de admisión al hospital.

Patient Age, Gender, Race: Atributos demográficos básicos.

Department Referral: Departamento al que fue referido el paciente (e.g., General Practice, Orthopedics).

Admission Flag: Indica si el ingreso fue efectivo (TRUE) o no (FALSE).

Satisfaction Score: Puntuación de satisfacción del paciente (escala 1–10).

Waittime: Tiempo de espera antes de ser atendido (en minutos).

Patients CM: Campo binario que podría representar condición médica crónica, complicación o categorización especial (requiere análisis adicional).

**Este dataset permite realizar análisis exploratorios enfocados en:**

Experiencia y satisfacción del paciente.

Tiempos de espera por departamento.

Tendencias de admisión según edad, género o etnia.

Comparación entre ingresos efectivos y no efectivos.
